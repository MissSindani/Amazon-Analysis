```python
# importing necessary libraries for our data analysis
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime

%matplotlib inline
```

```python
raw_data = pd.read_csv('amazon_prime_users.csv')
raw_data.head()
```

```python
data_copy = raw_data.copy()
```

```python
data_copy.drop(['User ID', 'Username', 'Email Address'], axis=1, inplace=True)
```

```python
data_copy.isnull().sum()
```

```python
data_copy['Location'].value_counts()
```

### What are the top-selling categories?

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Favorite Genres', data=data_copy, color='brown')
plt.show()
```

#### Average Ratings On Our Platform

```python
sns.countplot(y = data_copy['Feedback/Ratings'], color='brown')
plt.xlabel('How Many People Rated It')
plt.ylabel('Rating')
plt.show()
```

#### How long do users stay subscribed?

```python
sns.countplot(x = data_copy['Subscription Plan'], color='brown')
plt.show()
```

#### What devices are most commonly used to access the service?

```python
sns.countplot(x = data_copy['Devices Used'], color='brown')
plt.xlabel('Number Of Access')
plt.show()
```

#### How frequently do people engage with the platform?

```python
sns.countplot(x = data_copy['Engagement Metrics'], color='brown')
plt.xlabel('Customer Engagement')
plt.ylabel('Number Of People')
plt.title('Number Of People Engaging On The Platform')
plt.show()
```
